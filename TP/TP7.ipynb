```python
try:
    import cv2
except Exception as  e:
    print(f"erreur {e}")
import numpy as np
import time 
import math
```

```python
cap = cv2.VideoCapture(1)
fram_width = int(cap.get(3))
fram_height = int(cap.get(4))
```

```python
fourcc = cv2.VideoWriter_fourcc('X','V','I','D')
# fps => frame per sconde dans ce cas =50 
# le temps d'echantaillonage waitkey(20)
out = cv2.VideoWriter('output_out.avi',fourcc , 50, (fram_width,fram_height))
```

```python
if not cap.isOpened():
    print("error  capture")
    exit(0)
    
```

```python
while (cap.isOpened):
    ret , frame = cap.read()
    t_start = time.time()
    if not ret :
        print("error read frame")
        break
    frame = cv2.flip(frame,0)#=> verser l'image 1/0/
    out.write(frame)
    cv2.imshow("image", frame)
    if cv2.waitKey(20) & 0xFF == ord('q'):
        break
    print("time:" ,1/(time.time()-t_start))
out.release()
cv2.destroyAllWindows()
```
